'''Создать 5 двумерных массивов разной формы из массива a. Не использовать в аргументах метода reshape число -1.

```python
a1 = np.reshape(a[0:4], (2,2))
a2 = np.reshape(a[1:7], (2,3))
a3 = np.reshape(a[0:9], (3,3))
a4 = np.reshape(a[3:9], (2,3))
a5 = np.reshape(a[0:7:2], (2,2))
print(a1, '\n', a2, '\n', a3, '\n', a4, '\n', a5)
```

```python
a1 = np.reshape(a[0:4], (2,-1))
a2 = np.reshape(a[1:7], (2,-1))
a3 = np.reshape(a[0:9], (3,-1))
a4 = np.reshape(a[3:9], (-1,3))
a5 = np.reshape(a[0:7:2], (-1,2))
print(a1, '\n', a2, '\n', a3, '\n', a4, '\n', a5)
```
